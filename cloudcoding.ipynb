### **Step 4. BIG-IP Data Acquisition**
This is where things starts to get interesting. Using F5 SDK and IPStack modules function, you can now obtain BIG-IP's device version and location using its IP address or hostname. 

Using "[for loop](https://www.educba.com/for-loop-in-python/)" method in Python, we group the data collected and construct a list. 

flowchartpng.png

---




```
from ipstack import GeoLookup # Import GeoLookup function to lookup device location based on IP addresses

geo_lookup = GeoLookup("0cbeddca8c7f050ead7d6eff52a73bb4") # States the IPstack API token used for authentication

iplist = ["54.151.170.57", "34.205.120.66", "3.121.15.40"] #IP list of BIG-IPs
data = []  #Initiate a variable to compile data

for ip in iplist:
  geoinfo = geo_lookup.get_location(ip) #Lookup Location
  location = geoinfo['city'] #Parse the lookup response and extract only 'city' data
  device = ManagementClient(ip, user='admin', password='PythoN@3') #Connect to BIG-IP based on IP provided
  version = device.get_info() #Obtain device version
  data.append([ip, location, version]) #Append collected info to the 'data' variable
  
print(data)
```

### **Step 5. Viewing Data In Table Format using Panda Dataframe Module**
The data collected above is presented in a single line. To build the collected data in table format, you can deploy data framework such as [Panda Dataframe](https://pandas.pydata.org/pandas-docs/stable/user_guide/dsintro.html) to manage the data collected. To do this, run the code below to insert the data into framework and name the columns for your data. 


```
import pandas as pd  # Import the panda package
df = pd.DataFrame(data, columns=["IP", "Location", "Device Version"])
print(df)
```
